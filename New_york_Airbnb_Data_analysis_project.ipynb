```python
# Importing the important libraries required for the data analysis and visualization of dataset
import pandas as pd 
import numpy as np 
import seaborn as sns    
import matplotlib.pyplot as plt      
```

```python
# assigining the dataset of airbnb into the variable name df using pandas library. 

df = pd.read_csv('/kaggle/input/new-york-city-airbnb-open-data/AB_NYC_2019.csv')      
```

```python
# Number of rows and columns in dataset. 

df.shape    
```

```python
# Seeing some top values of our dataset

df.head()      
```

```python
# Seeing stastical data of our dataset  

df.describe()    
```

```python
# Seeing information about our dataset , it informs us about RangeIndex , non_null values and their data types 

df.info()      
```

```python
# Check missing values  

df.isnull().sum()      
```

```python
# Dropping uneccessary values  
# latitude and longitude are the redudancies in my dataset . 
# last_review and reviews_per_month which have lost of 10052 something null values .

df.drop(['latitude' , 'longitude' , 'last_review' , 'reviews_per_month']  , axis = 1 , inplace = True)       
```

```python
# Displaying top rows for checking weather dropped columns are dropped or not .

df.head(10)         
```

```python
# dropping duplicate values  
df.drop_duplicates(inplace = True)  
df.shape  
```

**No duplicate values found**

```python
# Analyzing weather there outliers  in our prices of the rooms .
# we use Boxplot . 

plt.figure(figsize = [14 , 7]) 
sns.boxplot(df['price']) 
plt.show() 
```

# It is showing lot of outliers in the prices of the rooms because mean is around 153.00 and it's 75% of data is below 176.00 . 


```python
# Analyzing the outliers of prices according to the neighbourhood_group(areas)   . 
plt.figure(figsize = [14, 7]) 
sns.boxplot(x = df['neighbourhood_group'] , y = df['price']) 
plt.show()
```

```python
plt.figure(figsize = [14, 7]) 
sns.boxplot(x = df['room_type'] , y = df['price'])   
plt.show()
```

**We found same in both neighbourhood_group and room_type we are geeting outliers**

```python
# We can have to find area and host names on the basis of calculated_host_listing_count . 

host_areas_listings = df.groupby(['host_name' , 'neighbourhood_group'])['calculated_host_listings_count'].max().reset_index() 

host_areas_listings = host_areas_listings.sort_values(by = 'calculated_host_listings_count' , ascending = False).head(10) 

host_areas_listings 
```

```python
plt.figure(figsize = [14, 7]) 
sns.barplot(data = host_areas_listings , x = host_areas_listings['host_name'] , y = host_areas_listings['calculated_host_listings_count'] , hue = 'neighbourhood_group') 
plt.show()  
```

**Bar plot is showing the top 10 area wise calculed listing count of hosts Sonder(NYC) has highest listing in Manhattan area followed by Blueground . ***

```python
# Now we are going to get someithing from room types and their areas 
# Average prices of the different room types in different areas . 

# Now we are going the maximum prices of the rooms according to their room type and neighbourhood 

room_type_counts_prices = df.groupby(['room_type','neighbourhood_group'])['price'].mean().reset_index()    
room_type_counts_prices = room_type_counts_prices.sort_values(by = 'price' , ascending = False)
room_type_counts_prices  
```

```python
plt.figure(figsize = [14 , 7]) 
sns.barplot(data = room_type_counts_prices , x = 'room_type' , y = 'price' , hue = 'neighbourhood_group')  
plt.title('Average prices according to the room types in different areas')
plt.show()   
```

**In this barplot we seen that average prices of the room types according to the area  , Entire home/apt has highest average prices.**

**Manhattan has highest averge prices in different room types.**

```python
plt.figure(figsize = [14 , 7]) 
sns.countplot(data = df , x = 'room_type' , hue = 'neighbourhood_group' ) 
plt.title(' Number of rooms available into the different areas and their room types') 
plt.show()
```

**In this barplot we are getting information about the availability of the rooms and their types in different regions**

```python
plt.figure(figsize = [14 , 7]) 
sns.scatterplot(x = df['price'] , y = df['number_of_reviews'])  
plt.title('Number of reviews VS price') 
plt.show() 
```

**We get the information that number of reviews are greater where the price is less **

```python
# We are seeing here about the maximum reviews for the particular areas 
area_wise_reviews = df.groupby('neighbourhood_group')['number_of_reviews'].count().reset_index() 
area_wise_reviews = area_wise_reviews.sort_values(by = 'number_of_reviews' , ascending = False) 
area_wise_reviews  
```

**We trying to get the information about area wise number of reviews given**

```python
plt.figure(figsize = [14 , 7]) 
plt.pie(labels = area_wise_reviews['neighbourhood_group'] , x = area_wise_reviews['number_of_reviews'])  
plt.title('Number of reviews according to neighbourhood_group') 
plt.show()  
```

**We are getting the information about Manhattan has the highest number of counts for availablity of the rooms**

```python
# Now we are going to check which host is bussiest or has maximum number of number of reviews for his romms 
bussy_hosts = df.groupby(['host_id','host_name','room_type', 'neighbourhood_group'])['number_of_reviews'].max().reset_index() 
bussy_hosts = bussy_hosts.sort_values(by = 'number_of_reviews' , ascending = False).head(10) 
bussy_hosts  
```

```python
plt.figure(figsize = [14 , 7]) 
sns.barplot(data = bussy_hosts , x = bussy_hosts['host_name'] , y = bussy_hosts['number_of_reviews'] , hue = 'room_type')  
plt.title('Top 10 Bussiest hosts with their room type') 
plt.show() 
```

**We found that Dona is the bussiest host because it has maximum number of reviews and private rooms are mostly booked**

```python
plt.figure(figsize = [14 , 7])
sns.boxplot(df['number_of_reviews'])  
plt.show()  
```

```python
crowd_areas = df.groupby(['neighbourhood_group' , 'room_type'])['minimum_nights'].count().reset_index() 
crowd_areas = crowd_areas.sort_values( by = 'minimum_nights' , ascending = False).head(10) 
crowd_areas 
```

```python
plt.figure(figsize = [14 , 7]) 
sns.barplot(data = crowd_areas , x = crowd_areas['room_type'] , y = crowd_areas['minimum_nights'] , hue = 'neighbourhood_group')  
plt.title('Counts of minimum nights with different room types and their neighbourhood_group') 
plt.show() 
```

**In this barplot we found that people like to stay at entire home and private rooms present in Manhattan , Brooklyn , & Queens. **
